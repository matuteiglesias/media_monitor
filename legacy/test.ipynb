```python
from serpapi import GoogleSearch

def fetch_google_news_bulk(queries, api_key, lang="en", region="us"):
    """
    Fetches Google News articles for multiple queries.
    
    Args:
        queries (list): List of keywords or phrases to search for.
        api_key (str): Your SerpAPI key.
        lang (str): Language for the search (default: "en").
        region (str): Region for the search (default: "us").
    
    Returns:
        dict: Dictionary where keys are queries and values are lists of articles.
    """
    all_articles = {}

    for query in queries:
        print(f"Fetching news for query: '{query}'")
        search = GoogleSearch({
            "q": query,
            "tbm": "nws",  # Target the News tab
            "hl": lang,    # Language
            "gl": region,  # Geographic region
            "api_key": api_key
        })

        results = search.get_dict()
        articles = results.get("news_results", [])

        # Process and save articles for this query
        formatted_articles = [
            {
                "title": article.get("title", ""),
                "link": article.get("link", ""),
                "snippet": article.get("snippet", ""),
                "source": article.get("source", ""),
                "date": article.get("date", "")
            }
            for article in articles
        ]
        all_articles[query] = formatted_articles

        # Print fetched articles
        for article in formatted_articles:
            print(f"Title: {article['title']}")
            print(f"Link: {article['link']}")
            print(f"Snippet: {article['snippet']}")
            print(f"Source: {article['source']}")
            print(f"Date: {article['date']}")
            print("---")
    
    return all_articles


# Replace with your SerpAPI key
API_KEY = "1e192ec3d6d4215d45e80146fe6ed05f6504f6d3ad9ce28c392cb9c79cdf95a4"
queries = ["macroeconomy", "inflation", "central bank"]
articles_by_query = fetch_google_news_bulk(queries, API_KEY)

# Acceder a los resultados por palabra clave
for query, articles in articles_by_query.items():
    print(f"\nArticles for query: {query}")
    for article in articles:
        print(f"- {article['title']}")
```
